```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("BTC-USD.csv")
```

```python
df.dtypes
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
closed_data = df["Close"]
closed_data.head()
```

```python
df = df[["Date","Close"]]
df.head()
```

```python
df["Close"].dropna(inplace=True)
```

```python
import datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%d.%m.%Y')
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
df['day'] = pd.DatetimeIndex(df['Date']).day
df['dayofyear'] = pd.DatetimeIndex(df['Date']).dayofyear
df['weekofyear'] = pd.DatetimeIndex(df['Date']).weekofyear
df['weekday'] = pd.DatetimeIndex(df['Date']).weekday
df['quarter'] = pd.DatetimeIndex(df['Date']).quarter
df['is_month_start'] = pd.DatetimeIndex(df['Date']).is_month_start
df['is_month_end'] = pd.DatetimeIndex(df['Date']).is_month_end
print(df.info())
```

```python
X = df["Date"].values
Y = df["Close"].values
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(16,8))
plt.title("stock prices",fontsize=16)
plt.xlabel("Date")
plt.ylabel("close values",fontsize=16)
plt.plot(X,Y)
plt.show()
```

```python
import sklearn.model_selection 
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)


```
